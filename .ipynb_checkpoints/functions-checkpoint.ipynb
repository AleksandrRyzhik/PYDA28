# Домашнее задание к лекции "Функции"

Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

1. перечень всех документов
```
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```
2. перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
```
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Общие требования к программе:
- код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
- в коде отсутствуют глобальные переменные (за исключением **documents** и **directories**);
- пользовательский ввод обрабатывается в цикле **while** до тех пор, пока пользователь явно не завершит программу (вводом команды "**q**").

````python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
````

````python
def doc_owner():
    user_input = input('Введите номер документа: ')
    result = 0
    for doc in documents:            
        if doc['number'] == user_input:
            result = 1
            print(f"Владелец документа: {doc['name']}")
            break
    if result == 0:
        print(f"Документ не найден в базе")
````

````python
def doc_directories(user_input):
    result = 0
    a = []
    for shelf, doc_list in directories.items():            
        if user_input in doc_list:
            result = 1
            a = [f"Документ хранится на полке: {shelf}", shelf]
            break
    if result == 0:
        a = [f"Документ не найден в базе", 0]
    return a
````

````python
def full_info():
    for document_info in documents:
        print(f"№: {document_info['number']}, тип: {document_info['type']}, \
владелец: {document_info['name']}, \
полка хранения: {doc_directories(document_info['number'])[1]}")
````

````python
def shelves_list():
    shelves_list = f"Текущий перечень полок: \
{', '.join(list(directories.keys()))}"
    return shelves_list
````

````python
def add_shelf():
    shelf_number = input('Введите номер полки: ')
    if shelf_number in directories:
        print(f"Такая полка уже существует. {shelves_list()}")
    else:
        directories.setdefault(shelf_number, [])
        print(f"Полка добавлена. {shelves_list()}")
````

````python
def delete_shelf():
    shelf_number = input('Введите номер полки: ')
    if shelf_number in directories:
        if directories[shelf_number] == []:
            del (directories[shelf_number])
            print (f"Полка удалена. {shelves_list()}")
        else:
            print (f"На полке есть документы, \
удалите их перед удалением полки. \
{shelves_list()}")
    else:
        print (f"Такой полки не существует. {shelves_list()}")
````

````python
def secretary_program():
    user_input = 'x'
    
    while user_input != 'q':
        user_input = input('Введите команду: ')
        if user_input == 'p':
            doc_owner()
        elif user_input == 's':
            print (doc_directories(user_input = \
input('Введите номер документа: '))[0])
        elif user_input == 'l':
            full_info()
        elif user_input == 'ads':
            add_shelf()
        elif user_input == 'ds':
            delete_shelf()
secretary_program()
````
